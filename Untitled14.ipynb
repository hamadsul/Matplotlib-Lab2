# Q1: Import matplotlib library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Q2: Create a one line chart using two arrays. The first array represents x and the second array represents y

```python
# Define the x and y arrays
x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.array([1, 2, 3, 5, 8, 13, 20])

# Create the line chart
plt.plot(x, y)

# Add labels and title
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('X and Y Arrays')

# Show the chart
plt.show()
```

# Q3: Create a two lines chart using the same previous arrays.
- But this time:

    - Mark each point with a circle.
    - Use dashed lines.
    - Set the size of the marker (circle) to size 10.
    - Set the width of the line to 3.
    - Add legends. 'Normal' for array x and 'Fast' for array y.

```python
x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.array([1, 2, 3, 5, 8, 13, 20])

plt.plot(x, y, 'o--', linewidth=3, markersize=10, label='Fast')
plt.plot(x, x, 'o--', linewidth=3, markersize=10, label='Normal')

plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('X and Y Arrays')

plt.legend()

plt.show()
```

# Q4: Display multiple line charts in one row using the charts given below.
- Give each chart a title
- Add a super title 'my data visulaization assignment'.

```python
# Define the data for the first chart
x_plot1 = np.array([1, 2, 3, 4, 5, 6, 7])
y_plot1 = np.array([1, 1, 2, 3, 5, 8, 13])

# Define the data for the second chart
x_plot2 = np.array([0, 1, 2, 3, 4, 5, 6])
y_plot2 = np.array([2, 4, 6, 8, 10, 12, 14])

# Define the data for the third chart
x_plot3 = np.array([0, 1, 3, 4])
y_plot3 = np.array([4, 6, 3, 4])

# Create a figure with 1 row and 3 columns
fig, axs = plt.subplots(1, 3, figsize=(15, 5))


# Create the first chart
axs[0].plot(x_plot1, y_plot1)
axs[0].set_title('First Chart')

# Create the second chart
axs[1].plot(x_plot2, y_plot2)
axs[1].set_title('Second Chart')

# Create the third chart
axs[2].plot(x_plot3, y_plot3)
axs[2].set_title('Third Chart')

```

# Q5: Define an array that has a values of your top three TV series and define another array to hold the rating of the three TV series out of 10. Create a bar chart using the two arrays
- Set a new color for the bars.
- Change the width of the bars to width 0.5.

```python
# Define the array of TV series and their ratings
tv_series = np.array(['Breaking Bad', 'Game of Thrones', 'Friends'])
ratings = np.array([7, 8, 10])

# Define the color for the bars
color = 'orange'

# Create the bar chart
plt.bar(tv_series, ratings, color=color, width=0.5)

# Add labels and title
plt.xlabel('TV Series')
plt.ylabel('Ratings out of 10')
plt.title('Top Three TV Series and Their Ratings')

# Show the chart
plt.show()
```

# Q6: Create a scatter plot using two arrays. The first array represents carAge and the second array represents carSpeed, set a specific color for each dot by using an array of different colors and add labels for the plot

- Set a new color for each dot by using an array of different colors.
- Change the size of the dots to size 350.
- Add labels for x-axis and y-axis.
- Adjust the transparency of the dots with the alpha argument. let the alpha value be 0.4

```python
# Define the data for the scatter plot
carAge = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11])
carSpeed = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78])
colors = np.array(['red', 'green', 'blue', 'purple', 'orange', 'pink', 'brown', 'gray', 'black', 'yellow'])

# Create the scatter plot
plt.scatter(carAge, carSpeed, s=350, c=colors, alpha=0.4)

# Add labels and title
plt.xlabel('Car Age')
plt.ylabel('Car Speed')
plt.title('Car Age vs Car Speed')

# Show the chart
plt.show()
```

# Q7: Create a pie chart using two arrays. The first array represents python_libraries and the second array represents your skills out of 10 at each library

- Set a new color for each wedge by using an array of different colors.
- Pull your favorite library wedge 0.3 from the center of the pie.
- Add a shadow to the pie chart

```python
# Define the data for the pie chart
python_libraries = np.array(['NumPy', 'Pandas', 'Matplotlib', 'Seaborn', 'Plotly', 'Scikit-learn'])
skills = np.array([6, 5, 10, 1, 1, 1])
colors = np.array(['red', 'green', 'blue', 'purple', 'orange', 'pink'])

# Pull the first wedge out from the center of the pie
explode = (0, 0, 0.3, 0, 0, 0)

# Create the pie chart
plt.pie(skills, labels=python_libraries, colors=colors, explode=explode, shadow=True)

# Add a title
plt.title('My Skills with Python Libraries')

# Show the chart
plt.show()
```

# Q8: Create a histogram with 250 values and 15 bars

```python
# Generate 250 random values
values = np.random.randn(250)

# Create the histogram
plt.hist(values, bins=15)

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of 250 Random Values with 15 Bars')

# Show the chart
plt.show()
```
